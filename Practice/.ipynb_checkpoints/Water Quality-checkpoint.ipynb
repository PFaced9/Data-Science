```python
X = data.drop(columns='is_safe', axis = 1)
y = data['is_safe']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y = pd.Series(y)
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
# y_scaled = sc.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=55, n_jobs=-1, max_features=7, max_depth=10, random_state=42)
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
from sklearn.model_selection import cross_val_score
score = cross_val_score(rfc,X,y, cv=10,n_jobs=-1,scoring='accuracy')
```

```python
np.mean(score)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
params = {
    'max_depth' : [None,1,2,4,8,10,9,3,5,7],
    'min_samples_split' : [1,2,4, 10],
    'min_samples_leaf' : [1, 2],
    'max_features' : ['auto','sqrt','log2'],
    'criterion' : ['gini','entropy']
}


dtree = DecisionTreeClassifier()
grid_search = GridSearchCV(estimator=dtree,
                           param_grid=params,
                           cv=5,
                           scoring='accuracy',
                           n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_score_)
print(grid_search.best_params_)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred2 = lr.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
score = cross_val_score(lr,X,y, cv=10,n_jobs=-1,scoring='accuracy')
```

```python
np.mean(score)
```
