## Basic Graph

```python
plt.plot([1,2,3],[2,4,6])        # x=1,2,3    y=2,4,6
```

```python
#or you can also write the above graph in this type always the size of x and y should be the same
x=[0,1,2,3,4]
y=[0,3,6,9,12]

#resize your graph    dpi gives you the high quality of graph
plt.figure(figsize=(6,3),dpi=100)     #dpi stands for dots per inches it gives the pixels of figure the default dpi is 100

#plt.plot(x,y , label="stock market",color="green",linewidth =3,linestyle="--",marker="*",markersize=10,markeredgecolor="red")
#short hand notation
#fmt=[color,marker,line,etc]    it should be in order wise

plt.plot(x,y,"r^--",label="stock market")  #{r=red,(^)=marker "--"=linestyle}<- short hand notation

#fontdict is not mandatory if you want to change some font details you want to use it

#you can mention the title 
z=plt.title("Line graph",fontdict={"fontname":"Times New Roman","fontsize":20}) 
plt.xlabel("x Axis",fontdict={"fontname":"comic sans ms","fontsize":15})
plt.ylabel("Y Axis",fontdict={"fontname":"comic sans ms","fontsize":15})

#if you want only integer values on x rather than decimal values
# graph automatically resizes based on how we set the scale
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,4,8,12,16,20,24])



#line 2

#select interval we want to plot points at blue color
x2=np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:7],x2[:7]**2,"b",linewidth=2.5,label="share peak line")       #it takes 2 values x,y where x=x2[0:7]and y=x2[0:7]**2

# we are making some changes with the same blue line by showing the cutoff line
plt.plot(x2[6:],x2[6:]**2,linestyle="--",color="g",label="safe zone")

#legend is nothing but mentoining the lable, can see at the right top at the graph
plt.legend()

# adding grid to the page

plt.grid(linestyle='--',linewidth=0.5,color='black')

#save the graph  you can change the graph dpi while saving it consume less space 

plt.savefig("matplotlib_mygraph",dpi=100)

#if we you want to print the graph we should print plt.show()
plt.show()
```

## Bar chart

```python
lables=["A","B","c"]
values=[1,4,2]
bars=plt.bar(lables,values)

# Hatching
#1st way
#bars[0].set_hatch("/")
#bars[1].set_hatch("o")
#bars[2].set_hatch("*")

#you can do this in other way 

#2nd way
patterns=["/","o","*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))   # here where it removes the list one by one so it calls the pattern and make use of it

plt.figure(figsize=(6,4))


plt.show()
```

## Pie Charts

### Q1] Get the left foot and right foot players and draw the pie chart based on that 

```python
left_foot=fifa.loc[fifa['Preferred Foot'] == "Left"].count()[0]
print("left_foot =",left_foot)
right_foot=fifa.loc[fifa["Preferred Foot"]=="Right"].count()[0]
print("right_foot =",right_foot)

labels=["left_foot players","right_foot players"]

plt.pie([left_foot,right_foot],labels=labels,autopct="%.2f  %%")        #autopct=autopercentage and the max decimal=2  so"%.2f"

plt.title("Foot Preference of FIFA Players",fontdict={"fontname":"times new roman","fontsize":25})




plt.show()

```

## Q2] By using the weight of the football players draw the Pie chart 
where min weight <125,light_medium_weight>=125&<150,medium weight >=150&<175,medium heavy weight>=175&<200 heavy weight>=200

```python
fifa["Weight"]    # The weights are showing in string format lets convert in to integer
```

```python
fifa.Weight=[int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]    # Using list comprension-> using strip will remove first and last strip value which are present in data  and making to print only intger values

plt.style.use("default")              #there are many different type of style to print the pie chart

light=fifa.loc[(fifa["Weight"]<125)].count()[0]
light_medium=fifa.loc[(fifa["Weight"]>=125) & (fifa["Weight"]<150)].count()[0]
medium=fifa.loc[(fifa["Weight"]>=150) &(fifa["Weight"]<175)].count()[0]
medium_heavy=fifa.loc[(fifa["Weight"]>=175) & (fifa["Weight"]<200)].count()[0]
heavy=fifa.loc[(fifa["Weight"]>=200)].count()[0]     

weights =[light,light_medium,medium,medium_heavy,heavy]
labels=["under 125","125-150","150-175","175-200","over 200"]

explode=[0.4,.2,0,0,.4]          # The distance from the center point to explode

plt.pie(weights,labels=labels,autopct="%.2f  %%",pctdistance=0.8, explode=explode)             #pct= percent distance from the center of the graph, inside the circle(graph) is considered as 1 
                                                                                            #explode break the circle and openup the graph and prints it, It make visible when their is very less percentage of values
    
plt.title("Weight Distribution of FIFA Players",fontdict={"fontname":"times new roman","fontsize":30})    
plt.show()
```

## boxplot

```python
plt.style.use("default")

plt.figure(figsize=(8,6))

barcelona=fifa.loc[fifa.Club=="FC Barcelona"]["Overall"]
madrid=fifa.loc[fifa.Club=="Real Madrid"]["Overall"]
revs=fifa.loc[fifa.Club=="New England Revolution"]["Overall"]

labels=["FC Barcelona","Real Madrid","NEw England Revolution"]

boxes=plt.boxplot([barcelona,madrid,revs],labels=labels,patch_artist = True,medianprops={"linewidth":3})

for box in boxes["boxes"]:
    # set edge color
    box.set(color="blue",linewidth=3)
        
    #change fill color  
    box.set(facecolor="grey")
        
plt.title("Professional Football Team Comparson")
plt.ylabel("FIFA Overall Rating")

plt.show()
```
