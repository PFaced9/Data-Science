## Aggregate Statistics (Groupby)

```python
#adding up all the data grouping by type 1 
df.groupby(["Type 1"]).sum()

#it gives you sum of all the value and grouped by type1
#ex adding all the values of if we have bug 5 count it will add all(bug) 5 total and gives you the single output
```

```python
#it gives you the mean (mean=average)
df.groupby(["Type 1"]).mean()
```

```python
df['Type 1'].value_counts().sort_index()
```

```python
# if we want to count no of data in type 1
df["count"]=1
df.groupby(["Type 1"]).count()["count"]
```

```python
#let count in type 1(bug) what the data present in Type 2
df["count"]=1
df.groupby(["Type 1","Type 2"]).count()["count"]
```

```python
#it will give you the average of everything grouping by defense
#where it take the type1(value) and make the average of all the values and we are going to sort by Defense and we are printing in descending order

df.groupby(["Type 1"]).mean().sort_values("Defense",ascending =False)
```

```python
df.groupby(["Type 2"]).mean().sort_values("Attack",ascending=False)
```

## working with large amounts of data

```python
 #printing every data by seperating with 5 rows where it helps to read easily we have have large amount of data we use chunksize todo this operation
for df in pd.read_csv("modified.csv",chunksize=5):
    print("chunk df")
    print(df)
```

```python
new_df=pd.DataFrame(columns=df.columns)
for df in pd.read_csv("modified.csv",chunksize=5):
    results=df.groupby(["Type 1"]).count()
    new_df=pd.concat([new_df,results])
print(new_df)
```
