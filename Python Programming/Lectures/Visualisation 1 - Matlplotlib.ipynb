### Line Plot

```python
month = np.arange(1,13)
prices = [101,112,203,304,335,406,507,608,709,810,795,854]
plt.plot(month, prices, color='b', marker='*')
plt.title('Price Trend for a Year')
plt.xlabel('Month')
plt.ylabel('Price')
plt.grid(linestyle='dashdot', linewidth=0.5, color='red')
plt.show()
```

### Multiline Plot

```python
x = [7, 14, 21, 28, 35, 42, 49]
y = [5, 12, 19, 21, 31, 27, 35]
z = [3, 5, 11, 20, 15, 29, 31]
plt.plot(x,y, color='b')
plt.plot(x,z, color='r')
plt.plot(y,z, color='g')
plt.show()
```

### Subplots

```python
fig,ax = plt.subplots(2,2)
x = np.linspace(0,8,1000)
ax[0,0].plot(x,x**2,'g')
ax[0,1].plot(x,x**3,'r')
ax[1,0].plot(x,np.sin(x),'y')
ax[1,1].plot(x,np.cos(x),'blue')
ax[0,0].grid(linestyle='--', linewidth=0.6, color='orange')
ax[1,0].grid(linestyle='--', linewidth=0.6, color='orange')
ax[1,1].grid(linestyle='--', linewidth=0.6, color='orange')
ax[0,1].grid(linestyle='--', linewidth=0.6, color='orange')

plt.show()
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

### Histograms

```python
plt.hist(penguins.bill_depth_mm,bins=20, color='orange')
plt.show()
```

### Boxplot

```python
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
# plt.boxplot(data)
plt.boxplot(data, showfliers=False)
plt.show()
```

### Barplot

```python
penguins['species'].value_counts().plot(kind='bar', color='b')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Species Distribution')
plt.show()
```

```python
penguins['island'].value_counts().plot(kind='barh', color='lightsalmon')
plt.xlabel('Count')
plt.ylabel('Islands')
plt.title('Island Distribution')
plt.show()
```

```python
plt.bar('species','flipper_length_mm', data=penguins, color='teal')
plt.show()
```

```python
Python_marks = (50,65,40,35,77)
R_marks = (55,72,94,70,85)
index = np.arange(5)
plt.bar(x=index, height=Python_marks, width=0.35, label='Python')
plt.bar(x=index+0.35, height=R_marks, width=0.35, label='R')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Scores by students')
plt.legend()
plt.xticks(ticks=index+0.35/2, labels=('A','B','C','D','E'))
plt.show()
```

#### Stacked Barplot

```python
Python_marks = (50,65,40,35,77)
R_marks = (55,72,94,70,85)
index = np.arange(5)
plt.bar(x=index, height=Python_marks, label='Python')
plt.bar(x=index, height=R_marks, bottom=Python_marks, label='R')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Scores by students')
plt.legend()
plt.xticks(ticks=index+0.35/2, labels=('A','B','C','D','E'))
plt.show()
```

### Piechart

```python
plt.pie(penguins['species'].value_counts(), labels=penguins['species'].unique(), autopct='%1.1f%%')
plt.title('Species Distribution')
plt.show()
```

```python
plt.pie(penguins['species'].value_counts(), labels=penguins['species'].unique(), autopct='%1.1f%%', colors=['orchid', 'gold', 'green'])
plt.title('Species Distribution')
plt.show()
```

```python
explode = [0.1,0,0]
plt.pie(penguins['species'].value_counts(), labels=penguins['species'].unique(), autopct='%1.1f%%', colors=['tomato', 'cornflowerblue', 'orchid'], explode=explode)
plt.title('Species Distribution')
plt.show()
```

```python
plt.pie(penguins['species'].value_counts(), labels=penguins['species'].unique(), autopct='%1.1f%%')
plt.gcf()
circle = plt.Circle(xy=(0,0), radius=0.4, color='white')
plt.gca().add_artist(circle)
plt.title('Species Distribution')
plt.show()
```
