# LINE PLOT

```python
month = np.arange(1,13)
prices = [101, 112, 203, 304, 335, 406, 507, 608, 709, 810, 795, 854]

#plot prices vs. month
#'color assigns the color of the plot
#market assigns the shape of the data point

plt.plot(month, prices, color='darkgreen', marker='*')
# label axes
plt.xlabel("Month") 
plt.ylabel("Prices")
# title
plt.title("Prices trend of a Year")
#grid lines
plt.grid(linestyle='dashdot', linewidth='0.7', color='blue')
# show
plt.show()
```
