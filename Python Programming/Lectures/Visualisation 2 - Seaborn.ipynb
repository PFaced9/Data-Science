```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips)
```

```python
sns.stripplot(x="sex", y="tip", data=tips)
plt.title("Strip plot for Gender and Tips")
plt.show()
```

```python
# sns.displot(data=tips, x="total_bill", color="olive", kind='kde')
sns.displot(data=tips, x="total_bill", color="olive", kde='true')
plt.show()
```

```python
# sns.jointplot(data=tips, x='total_bill', y='tip', color='sandybrown')
# sns.jointplot(data=tips, x='total_bill', y='tip', color='sandybrown', kind='hex')
# sns.jointplot(data=tips, x='total_bill', y='tip', color='sandybrown', kind='kde')
sns.jointplot(data=tips, x='total_bill', y='tip', color='sandybrown', kind='reg')
plt.show()
```

```python
sns.violinplot(data=tips, y="total_bill", x="day", hue="smoker", split=True)
# sns.violinplot(data=tips, y="total_bill", x="day", hue="smoker", split=False)
plt.show()
```

```python
# sns.swarmplot(data=tips, x='sex', y='tip', color='red')
sns.swarmplot(data=tips, x='sex', y='tip', color='red', hue='smoker')
plt.show()
```

```python
# sns.pairplot(data=tips)
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
sns.countplot(data=tips, x="sex")
plt.text(x=-0.1, y=160, s=str(round(tips.sex.value_counts()[0]/len(tips)*100, 2))+'%')
plt.text(x=0.9, y=90, s=str(round(tips.sex.value_counts()[1]/len(tips)*100, 2))+'%')
plt.show()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
glue = sns.load_dataset("glue").pivot("Model", "Task", "Score")
sns.heatmap(glue, annot=True, fmt=".1f", linewidths=0.5, cmap="BuPu")
plt.show()
```
