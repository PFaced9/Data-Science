```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# OLS 
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
print("Value of Slope:",reg.coef_)
print("Value of B:", reg.intercept_)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='yellow')
plt.show()
```

```python
# LETS APPLY GRADIENT DESCENT ASSUMING SLOPE IS CONSTANT M = 78.33
# LETS ASSUME THE STARTING VALUE OF B = 0
y_pred = ((78.33 * x) + 0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red', label='OLS')
plt.plot(x,y_pred, color='#00a65a', label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
loss_slope = -2 * np.sum(y - m*x.ravel() -  b)
loss_slope
```

```python
# LETS TAKE THE LEARNING RATE  = 0.1
lr = 0.1
step_size = loss_slope * lr
step_size
```

```python
#calculate the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35*x) + b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x), color='red', label='OLS')
plt.plot(x,y_pred, color='#00a65a', label='b = 0')
plt.plot(x,y_pred1, color='black', label='b = new b')
plt.legend() 
plt.show()
```
